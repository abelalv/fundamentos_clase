### Como solucionar ecuaciones de primer grado con Python y la librería sympy

Vamos a solucionar una ecuación de primer grado con Python y la librería sympy. 
Tomemos el ejemplo de la ecuación 

$$3x + 2 = 11.$$

En Python podemos trabajar de forma simbólica para resolver este tipo de ecuaciones. Para ello, primero debemos importar la librería sympy y definir la variable simbólica $x$. Luego, definimos la ecuación y la resolvemos. 

    


```python
from sympy import Symbol, Eq, solve

# Definimos la variable simbólica x como real
x = Symbol('x', real=True)

# Definimos la ecuación

ecuacion = Eq(3*x + 2, 11)

# Resolvemos la ecuación
print("Resolviendo la ecuación")
display(ecuacion)
solucion = solve(ecuacion)

print("la solución de la ecuación es:",solucion)
```

Note que el programa imprime la solución  $[3]$. Esto significa que la solución de la ecuación es $x = 3$.

Observemos otro ejemplo, ahora con la ecuación

$$(x+5)(x-1)=x^2-1.$$
    

```python
from sympy import Symbol, Eq, solve

# Definimos la variable simbólica x
x = Symbol('x', real=True)

# Definimos la ecuación

ecuacion = Eq((x + 5)*(x-1), x**2-1)

# Resolvemos la ecuación
print("Resolviendo la ecuación")
display(ecuacion)
solucion = solve(ecuacion)

print("la solución de la ecuación es:",solucion)
```

Note que para la ecuación anterior $x=1$ es solución y esto es porque si se sustituye la solución en la ecuación original, la igualdad se cumple. Sin embargo, si intentamos reescribir la ecuación anterior factorizándola del siguiente modo:
$$(x-1)(x+5)=(x-1)(x+1),$$
se podría pensar que al dividir en ambos lados de la igualdad por $(x-1)$ se obtiene 

$$x+5=x+1,$$

lo cual es incorrecto en todos los números reales **¿Por qué?**.

La librería sympy también permite resolver inecuaciones de una forma muy general, para ello debemos ingresar la inecuación en la función solve_univariate_inequality, y este nos devolverá el intervalo de la solución. Por ejemplo, para la inecuación

$$3x+5>2,$$

```python
import sympy as sp

# Definir la variable
x = sp.symbols('x', real=True)

# Definir la inecuación
inequality = 3*x +5> 2

# Resolver la inecuación
solution = sp.solve_univariate_inequality(inequality, x, relational=False)


# Mostrar la solución
print("Solución sin relacional:", solution)

```

Observe otro ejemplo 

$$-3x+5 \leq2,$$

```python
import sympy as sp

# Definir la variable
x = sp.symbols('x', real=True)

# Definir la inecuación
inequality = -3*x+5 <=2

# Resolver la inecuación
solution = sp.solve_univariate_inequality(inequality, x, relational=False)


# Mostrar la solución
print("Solución sin relacional:", solution)

```

**Note** que la solución de la inecuación anterior es $x\geq1$  y  la solución exhibida por el programa es `Interval(1, oo)` lo que significa  que la solución incluye el número real  1. Ahora, si se quiere resolver la inecuación

$$-3x+5 <2,$$

usamos el programa

```python
import sympy as sp

# Definir la variable
x = sp.symbols('x', real=True)

# Definir la inecuación
inequality = -3*x+5 <2

# Resolver la inecuación
solution = sp.solve_univariate_inequality(inequality, x, relational=False)


# Mostrar la solución
print("Solución sin relacional:", solution)
```

se encuentra que la solución es $x>1$ y la solución exhibida por el programa es `Interval.open(1, oo)` lo que significa que la solución no incluye el número real 1.
