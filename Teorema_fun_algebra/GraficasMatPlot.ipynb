```python
import matplotlib.pyplot as plt
import numpy as np

# Crear una figura con un tamao especfico
plt.figure(figsize=(12, 6))  # Ancho: 12 pulgadas, Alto: 6 pulgadas

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Graficar los datos
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ejemplo de Grfica con Tamao Ajustado')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear una figura
plt.figure(figsize=(12, 6))

# Graficar los datos
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ejemplo de Grfica con Lmites de Ejes Ajustados')
plt.grid(True)

# Ajustar los lmites de los ejes
plt.xlim(2, 8)  # Limitar el eje x de 2 a 8
plt.ylim(-1, 1)  # Limitar el eje y de -1 a 1

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Crear una figura con un tamao especfico y una resolucin especfica
plt.figure(figsize=(12, 6), dpi=100)  # DPI: 100 puntos por pulgada

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Graficar los datos
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ejemplo de Grfica con Resolucin Ajustada')
plt.grid(True)
plt.show()
```
