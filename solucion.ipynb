# Guía de Ejercicios de Proporcionalidad  

## Ejercicio 1: Proporcionalidad Directa con Fracciones
**Enunciado:** Si 3/4 de litro de pintura cubre 2/5 de una pared, ¿cuántos litros de pintura se necesitan para cubrir toda la pared?

**Solución:**
$$
\text{Litros necesarios} = \frac{3/4}{2/5} = \frac{3/4 \times 5/2} = \frac{15}{8} = 1.875 \text{ litros}
$$

## Ejercicio 2: Proporcionalidad Inversa con Velocidades
**Enunciado:** Un avión tarda 6 horas en recorrer una distancia a una velocidad de 800 km/h. ¿Qué velocidad debe mantener para recorrer la misma distancia en 4 horas?

**Solución:**
$$
\text{Velocidad} = \frac{\text{Distancia}}{\text{Tiempo}} = \frac{800 \text{ km/h} \times 6 \text{ h}}{4 \text{ h}} = 1200 \text{ km/h}
$$

## Ejercicio 3: Proporcionalidad Directa en Física
**Enunciado:** La intensidad de una fuerza (F) es directamente proporcional al área (A) sobre la que actúa, y F es 200 N cuando A es 5 m². ¿Cuál es el área cuando la fuerza es 500 N?

**Solución:**
$$
F \propto A \Rightarrow \frac{F_1}{A_1} = \frac{F_2}{A_2} \Rightarrow \frac{200}{5} = \frac{500}{A_2} \Rightarrow A_2 = \frac{500 \times 5}{200} = 12.5 \text{ m}²
$$

## Ejercicio 4: Proporcionalidad Compuesta con Temperaturas
**Enunciado:** En un experimento, la relación entre la temperatura T y el tiempo t es tal que T es directamente proporcional a t². Si T es 100°C cuando t es 5 minutos, ¿cuál será la temperatura cuando el tiempo sea 8 minutos?

**Solución:**
$$
T \propto t^2 \Rightarrow \frac{T_1}{t_1^2} = \frac{T_2}{t_2^2} \Rightarrow \frac{100}{5^2} = \frac{T_2}{8^2} \Rightarrow T_2 = 100 \times \frac{64}{25} = 256 \text{ °C}
$$

## Ejercicio 5: Proporcionalidad en Química con Reacciones
**Enunciado:** En una reacción química, se combinan dos sustancias A y B en la proporción de 3:7. Si se utilizan 42 gramos de B, ¿cuántos gramos de A se necesitan?

**Solución:**
$$
\frac{A}{B} = \frac{3}{7} \Rightarrow A = \frac{3}{7} \times 42 = 18 \text{ gramos}
$$

## Ejercicio 6: Proporcionalidad Directa e Inversa en Economía
**Enunciado:** Un fabricante produce x unidades de un producto a un costo total de C dólares. Si C es directamente proporcional a x y C es inversamente proporcional al tiempo t, ¿cómo cambia C cuando x se duplica y t se reduce a la mitad?

**Solución:**
$$
C \propto \frac{x}{t} \Rightarrow C \text{ nuevo} = k \times \frac{2x}{\frac{t}{2}} = k \times \frac{4x}{t} = 4C
$$

## Ejercicio 7: Proporcionalidad en la Creación de Modelos
**Enunciado:** Para construir un modelo a escala de un edificio, se utiliza una escala de 1:50. Si la altura real del edificio es de 100 metros, ¿cuál es la altura del modelo? Si el volumen del edificio es 500,000 m³, ¿cuál es el volumen del modelo?

**Solución:**
$$
\text{Altura del modelo} = \frac{100 \text{ m}}{50} = 2 \text{ m}
$$
$$
\text{Volumen del modelo} = \left(\frac{1}{50}\right)^3 \times 500,000 \text{ m}^3 = \frac{1}{125,000} \times 500,000 = 4 \text{ m}^3
$$

## Ejercicio 8: Proporcionalidad y Ecuaciones Diferenciales
**Enunciado:** La velocidad de descomposición de una sustancia es proporcional a la cantidad restante de la sustancia. Si la cantidad inicial es 100 gramos y la cantidad restante después de 10 horas es 60 gramos, ¿cuál será la cantidad restante después de 20 horas?

**Solución:**
$$
\text{Modelo exponencial: } N(t) = N_0 e^{-kt}
$$
$$
N(10) = 60 = 100 e^{-10k} \Rightarrow e^{-10k} = 0.6 \Rightarrow -10k = \ln(0.6) \Rightarrow k = -\frac{\ln(0.6)}{10}
$$
$$
N(20) = 100 e^{-20k} = 100 \left( e^{-10k} \right)^2 = 100 \times 0.6^2 = 36 \text{ gramos}
$$

## Ejercicio 9: Proporcionalidad y Movimiento Armónico
**Enunciado:** La elongación x de un resorte es directamente proporcional a la fuerza F aplicada, con una constante de proporcionalidad k. Si F es 20 N cuando x es 0.4 m, ¿cuál es la constante k? ¿Cuál será la elongación cuando la fuerza aplicada sea 50 N?

**Solución:**
$$
F = kx \Rightarrow k = \frac{F}{x} = \frac{20 \text{ N}}{0.4 \text{ m}} = 50 \text{ N/m}
$$
$$
x = \frac{F}{k} = \frac{50 \text{ N}}{50 \text{ N/m}} = 1 \text{ m}
$$

## Ejercicio 10: Proporcionalidad Compuesta en Dinámica de Fluidos
**Enunciado:** El caudal Q de un fluido que pasa a través de una tubería es directamente proporcional al diámetro D de la tubería al cuadrado y a la diferencia de presión P entre los extremos de la tubería, e inversamente proporcional a la longitud L de la tubería. Si Q es 200 L/min cuando D es 10 cm, P es 20 kPa, y L es 50 m, ¿cuál será el caudal cuando D se incrementa a 15 cm, P se reduce a 15 kPa, y L se incrementa a 60 m?

$$
Q \propto \frac{D^2 P}{L}

Q_1 = k \frac{D_1^2 P_1}{L_1} = 200

# Guía de Ejercicios de Porcentajes Complejos - Soluciones

## Ejercicio 1: Incremento y Descuento Combinado
**Enunciado:** Un artículo tiene un precio inicial de $150. Primero se aplica un descuento del 20% y luego un aumento del 10%. ¿Cuál es el precio final del artículo?

**Solución:**
$$
\text{Precio después del descuento} = 150 \times (1 - 0.20) = 150 \times 0.80 = 120 \text{ dólares}
$$
$$
\text{Precio después del aumento} = 120 \times (1 + 0.10) = 120 \times 1.10 = 132 \text{ dólares}
$$

## Ejercicio 2: Interés Compuesto Anual
**Enunciado:** Si depositas $1,000 en una cuenta de ahorro con un interés compuesto anual del 5%, ¿cuál será el saldo después de 3 años?

**Solución:**
$$
A = P \left(1 + \frac{r}{n}\right)^{nt} = 1000 \left(1 + \frac{0.05}{1}\right)^{1 \times 3} = 1000 (1.05)^3 \approx 1157.63 \text{ dólares}
$$

## Ejercicio 3: Porcentaje de Mezcla
**Enunciado:** Tienes dos soluciones de salmuera. Una tiene un 10% de sal y la otra un 25% de sal. ¿Cuántos litros de cada solución necesitas mezclar para obtener 20 litros de una solución al 15% de sal?

**Solución:**
Sea \( x \) la cantidad de litros de la solución al 10% y \( y \) la cantidad de litros de la solución al 25%.
$$
x + y = 20 \\
0.10x + 0.25y = 0.15 \times 20 = 3 \\
\Rightarrow 10x + 25y = 30 \\
\Rightarrow 2x + 5y = 6
$$
Resolvemos el sistema de ecuaciones:
$$
x + y = 20 \\
2x + 5y = 6 \\
\Rightarrow x = 20 - y \\
2(20 - y) + 5y = 6 \\
40 - 2y + 5y = 6 \\
3y = -34 \\
y = 2 \\
x = 20 - 2 = 18
$$
Se necesitan 18 litros de la solución al 10% y 2 litros de la solución al 25%.

## Ejercicio 4: Cambios Percentuales Sucesivos
**Enunciado:** El precio de un producto aumenta un 15% en el primer año y disminuye un 10% en el segundo año. Si el precio inicial era de $500, ¿cuál es el precio después de dos años?

**Solución:**
$$
\text{Precio después del primer año} = 500 \times 1.15 = 575 \text{ dólares}
$$
$$
\text{Precio después del segundo año} = 575 \times 0.90 = 517.50 \text{ dólares}
$$

## Ejercicio 5: Comparación de Porcentajes
**Enunciado:** En una empresa, el 60% de los empleados son hombres y el resto mujeres. Si el 70% de los hombres y el 80% de las mujeres asistieron a una reunión, ¿qué porcentaje del total de empleados asistió a la reunión?

**Solución:**
Supongamos que hay 100 empleados.
$$
\text{Hombres que asistieron} = 0.60 \times 0.70 \times 100 = 42 \\
\text{Mujeres que asistieron} = 0.40 \times 0.80 \times 100 = 32 \\
\text{Total que asistieron} = 42 + 32 = 74 \\
\text{Porcentaje que asistió} = \frac{74}{100} \times 100 = 74\%
$$

## Ejercicio 6: Porcentaje de Aumento en Producción
**Enunciado:** Una fábrica produce 200 unidades de un producto en el primer mes. En el segundo mes, la producción aumenta un 25% respecto al primer mes. En el tercer mes, la producción aumenta un 20% respecto al segundo mes. ¿Cuál es la producción total en los tres meses?

**Solución:**
$$
\text{Producción segundo mes} = 200 \times 1.25 = 250 \text{ unidades} \\
\text{Producción tercer mes} = 250 \times 1.20 = 300 \text{ unidades} \\
\text{Producción total} = 200 + 250 + 300 = 750 \text{ unidades}
$$

## Ejercicio 7: Análisis de Pérdidas y Ganancias
**Enunciado:** Un comerciante compra un artículo por $250 y lo vende por $300. Luego, con el dinero obtenido, compra otro artículo y lo vende por $400. ¿Cuál es el porcentaje total de ganancia del comerciante?

**Solución:**
$$
\text{Ganancia en la primera transacción} = 300 - 250 = 50 \text{ dólares} \\
\text{Ganancia en la segunda transacción} = 400 - 300 = 100 \text{ dólares} \\
\text{Ganancia total} = 50 + 100 = 150 \text{ dólares} \\
\text{Porcentaje total de ganancia} = \frac{150}{250} \times 100 = 60\%
$$

## Ejercicio 8: Cálculo de Impuestos
**Enunciado:** Un producto tiene un precio neto de $500. Si se le aplica un impuesto del 18%, ¿cuál es el precio final del producto con el impuesto incluido?

**Solución:**
$$
\text{Precio final} = 500 \times 1.18 = 590 \text{ dólares}
$$

## Ejercicio 9: Distribución de Población
**Enunciado:** En una ciudad, el 30% de la población es menor de 18 años, el 50% tiene entre 18 y 60 años, y el resto tiene más de 60 años. Si la población total es de 120,000 personas, ¿cuántas personas tienen más de 60 años?

**Solución:**
$$
\text{Personas menores de 18 años} = 0.30 \times 120,000 = 36,000 \\
\text{Personas entre 18 y 60 años} = 0.50 \times 120,000 = 60,000 \\
\text{Personas mayores de 60 años} = 120,000 - 36,000 - 60,000 = 24,000
$$

## Ejercicio 10: Ajuste Salarial
**Enunciado:** El salario de un empleado se incrementa en un 10% cada año. Si su salario inicial es de $40,000, ¿cuál será su salario después de 4 años?

**Solución:**
$$
\text{Salario después de 1 año} = 40,000 \times 1.10 = 44,000 \text{ dólares} \\
\text{Salario después de 2 años} = 44,000 \times 1.10 = 48,400 \text{ dólares} \\
\text{Salario después de 3 años} = 48,400 \times 1.10 = 53,240 \text{ dólares} \\
\text{Salario después de 4 años} = 53,240 \times 1.10 = 58,564 \text{ dólares}
$$
