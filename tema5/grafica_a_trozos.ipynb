## grafica de funcion a trozos

```python

import matplotlib.pyplot as plt
import numpy as np

# Definimos una función a trozos utilizando la función piecewise de numpy
# Se definirá una función que sea cúbica en un intervalo y lineal en otro

x = np.linspace(-10, 10, 400)
y = np.piecewise(x, [x < -2, (x >= -2) & (x < 2), x >= 2], 
                 [lambda x: x**3 - 9*x, lambda x: x, lambda x: -x + 5])

# Graficamos la función a trozos
fig, ax = plt.subplots()
ax.plot(x, y, label=r'Función a trozos')

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)
ax.legend()
plt.title('Función a trozos')
plt.grid(True)
plt.show()
```
