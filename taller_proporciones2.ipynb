# Guía de Ejercicios de Proporcionalidad  
## Ejercicio 1: Proporcionalidad Directa con Fracciones
Si 3/4 de litro de pintura cubre 2/5 de una pared, ¿cuántos litros de pintura se necesitan para cubrir toda la pared?

## Ejercicio 2: Proporcionalidad Inversa con Velocidades
Un avión tarda 6 horas en recorrer una distancia a una velocidad de 800 km/h. ¿Qué velocidad debe mantener para recorrer la misma distancia en 4 horas?

## Ejercicio 3: Proporcionalidad Directa en Física
La intensidad de una fuerza (F) es directamente proporcional al área (A) sobre la que actúa, y F es 200 N cuando A es 5 m². ¿Cuál es el área cuando la fuerza es 500 N?

## Ejercicio 4: Proporcionalidad Compuesta con Temperaturas
En un experimento, la relación entre la temperatura T y el tiempo t es tal que T es directamente proporcional a t². Si T es 100°C cuando t es 5 minutos, ¿cuál será la temperatura cuando el tiempo sea 8 minutos?

## Ejercicio 5: Proporcionalidad en Química con Reacciones
En una reacción química, se combinan dos sustancias A y B en la proporción de 3:7. Si se utilizan 42 gramos de B, ¿cuántos gramos de A se necesitan?

## Ejercicio 6: Proporcionalidad Directa e Inversa en Economía
Un fabricante produce x unidades de un producto a un costo total de C dólares. Si C es directamente proporcional a x y C es inversamente proporcional al tiempo t, ¿cómo cambia C cuando x se duplica y t se reduce a la mitad?

## Ejercicio 7: Proporcionalidad en la Creación de Modelos
Para construir un modelo a escala de un edificio, se utiliza una escala de 1:50. Si la altura real del edificio es de 100 metros, ¿cuál es la altura del modelo? Si el volumen del edificio es 500,000 m³, ¿cuál es el volumen del modelo?

## Ejercicio 8: Proporcionalidad y Ecuaciones Diferenciales
La velocidad de descomposición de una sustancia es proporcional a la cantidad restante de la sustancia. Si la cantidad inicial es 100 gramos y la cantidad restante después de 10 horas es 60 gramos, ¿cuál será la cantidad restante después de 20 horas?

## Ejercicio 9: Proporcionalidad y Movimiento Armónico
La elongación x de un resorte es directamente proporcional a la fuerza F aplicada, con una constante de proporcionalidad k. Si F es 20 N cuando x es 0.4 m, ¿cuál es la constante k? ¿Cuál será la elongación cuando la fuerza aplicada sea 50 N?

## Ejercicio 10: Proporcionalidad Compuesta en Dinámica de Fluidos
El caudal Q de un fluido que pasa a través de una tubería es directamente proporcional al diámetro D de la tubería al cuadrado y a la diferencia de presión P entre los extremos de la tubería, e inversamente proporcional a la longitud L de la tubería. Si Q es 200 L/min cuando D es 10 cm, P es 20 kPa, y L es 50 m, ¿cuál será el caudal cuando D se incrementa a 15 cm, P se reduce a 15 kPa, y L se incrementa a 60 m?

# Guía de Ejercicios de Porcentajes  

## Ejercicio 1: Incremento y Descuento Combinado
Un artículo tiene un precio inicial de $150. Primero se aplica un descuento del 20% y luego un aumento del 10%. ¿Cuál es el precio final del artículo?



## Ejercicio 2: Porcentaje de Mezcla
Tienes dos soluciones de salmuera. Una tiene un 10% de sal y la otra un 25% de sal. ¿Cuántos litros de cada solución necesitas mezclar para obtener 20 litros de una solución al 15% de sal?

## Ejercicio 4: Cambios Percentuales Sucesivos
El precio de un producto aumenta un 15% en el primer año y disminuye un 10% en el segundo año. Si el precio inicial era de $500, ¿cuál es el precio después de dos años?

## Ejercicio 5: Comparación de Porcentajes
En una empresa, el 60% de los empleados son hombres y el resto mujeres. Si el 70% de los hombres y el 80% de las mujeres asistieron a una reunión, ¿qué porcentaje del total de empleados asistió a la reunión?

## Ejercicio 6: Porcentaje de Aumento en Producción
Una fábrica produce 200 unidades de un producto en el primer mes. En el segundo mes, la producción aumenta un 25% respecto al primer mes. En el tercer mes, la producción aumenta un 20% respecto al segundo mes. ¿Cuál es la producción total en los tres meses?

## Ejercicio 7: Análisis de Pérdidas y Ganancias
Un comerciante compra un artículo por $250 y lo vende por $300. Luego, con el dinero obtenido, compra otro artículo y lo vende por $400. ¿Cuál es el porcentaje total de ganancia del comerciante?

## Ejercicio 8: Cálculo de Impuestos
Un producto tiene un precio neto de $500. Si se le aplica un impuesto del 18%, ¿cuál es el precio final del producto con el impuesto incluido?

## Ejercicio 9: Distribución de Población
En una ciudad, el 30% de la población es menor de 18 años, el 50% tiene entre 18 y 60 años, y el resto tiene más de 60 años. Si la población total es de 120,000 personas, ¿cuántas personas tienen más de 60 años?

## Ejercicio 10: Ajuste Salarial
El salario de un empleado se incrementa en un 10% cada año. Si su salario inicial es de $40,000, ¿cuál será su salario después de 4 años?
