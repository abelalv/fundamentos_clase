## Trabajando con Python

### Operaciones simbólicas

Python permite trabajar con expresiones simbólicas y manipularlas de manera sencilla. Para ello se utiliza la librería `sympy`, que permite la manipulación de expresiones algebraicas. Inicialmente se debe importar  `sympy`, definir las variables simbólicas que se van a utilizar y luego realizar la operación que se busca resolver. Para mostrar los resultados de las operaciones se utiliza la función `display` de la librería `sympy`.

Empleando un caso del módulo pasado, los números racionales se pueden visualizar de manera más natural con la función `Rational`, esto es un objeto simbólico, no un número decimal.

Por ejemplo sumemos dos racionales:

$$\frac{1}{2} + \frac{1}{3}$$

```python
import sympy as sp

resultado=sp.Rational(1, 2) + sp.Rational(1, 3)

print("Resultado: ")

display(resultado)
```

### Operaciones con polinomios

El siguiente ejemplo muestra dos polinomios de segundo y tercer grado, respectivamente. Si desea especificar que las variables con las que va a operar pertenecen al conjunto de los números reales se utiliza el comando `real=True`.

```python
import sympy as sp

x = sp.symbols('x', real=True)

f = x**2 - 2*x - 3
g= 3*x**3 - 3*x**2 + 4*x - 5

display(f)
display(g)
```

#### Sumas y restas

Se pueden sumar o restar expresiones algebraicas, empleando el operador `+` y `-`, respectivamente. A continuación, se presentan ejemplos de suma y resta empleando las expresiones algebraicas `f` y `g`, previamente definidas.


```python
display(f+g)
display(g-f)
```

#### Multiplicación

La multiplicación de expresiones $f,g$ es decir $f\cdot g$ también se puede realizar de forma simbólica y para expandir la expresión resultante  se utiliza la función `expand` de sympy.


```python
display((f*g))
print("Expandiendo el resultado")
display(sp.expand(f*g))
```

#### Factorización

Del mismo modo,  `sympy` permite factorizar expresiones algebraicas. Para ello se utiliza la función `factor` de sympy. A continuación, se presenta la factorización del polinomio de segundo grado
 $x^2-2x-3$.

```python
f=x**2 - 2*x - 3
display(f)
print("Factorizado")
display(sp.factor(f))
```

```python
g=x**4-16
display(g)
print("Factorizado")
display(sp.factor(g))
```

#### Simplificación

Empleando `sympy` también es posible simplificar una expresión algebraica. Para ello se utiliza la función `simplify`. A continuación, se presenta un ejemplo de como se puede simplificar la siguiente expresión racional:
$$ \frac{x^2-4}{x^4-16}$$

```python
f=x**2 -4
g=x**4-16

display(f/g)
print("Simplificado")
display(sp.simplify(f/g))
```

En `sympy` puede usarse números racionales de forma exacta con la función `Rational`. A continuación, se presenta un ejemplo de como se puede usar la función `Rational` para usar un número racional en una expresión algebráica.

```python
display(x**2-x+sp.Rational(1,4))
print("Factorizado")
display(sp.factor(x**2-x+sp.Rational(1,4)))
```

```python
display(sp.factor(x**2/4-1))
```

`sympy` permite involucrar múltiples variables, por lo que es posible trabajar con nombres de variables distintas. A continuación, se presenta la expresión algebraica $x^2+2xy+y^2$ y se simplifica.

### Tanque rectangular para almacenamiento de agua potable

Para una población ubicada en una vereda del municipio de Cali, se diseñó un tanque para el almacenamiento de agua potable. Este tanque, de forma rectangular, tiene un volumen de $1315.7 m^3$, con un ancho de $11.6 m$ y un largo de $35 m$.

Se pide encontrar las siguientes dimensiones:
 1. La profundidad del tanque rectangular:

**Respuesta**

```python



import sympy as sp

# Definimos las variables
V, L, A, P = sp.symbols('V L A P')

# Dado el volumen, largo y ancho, definimos la ecuación
V_eq = sp.Eq(V, L * A * P)

# Despejamos la profundidad P
P_sol = sp.solve(V_eq, P)[0]

# Mostramos la ecuación resuelta simbólicamente
display(V_eq, P_sol)

# Sustituimos los valores conocidos
V_valor = 1315.7  # m^3
L_valor = 35      # m
A_valor = 11.6    # m

# Calculamos el valor numérico de la profundidad
P_valor = P_sol.subs({V: V_valor, L: L_valor, A: A_valor})

print("La profundidad, P = ",'{0:1.2e}'.format(P_valor), "m")
```

La función `solve` se utiliza para resolver ecuaciones. En este caso, está resolviendo la ecuación `V_eq` para la variable `P`. La función devuelve una lista de soluciones, y `[0]` al final de la línea de código está seleccionando la primera solución de esa lista. Si se quisiera considerar todas las soluciones, se podría eliminar el `[0]` y `P_sol` sería una lista de todas las soluciones posibles.

La función `subs` se utiliza para sustituir una variable simbólica por un valor o por otra expresión. En este caso, `P_sol.subs({V: V_valor, L: L_valor, A: A_valor})` está sustituyendo las variables simbólicas `V`, `L` y `A` en la expresión `P_sol` por los valores `V_valor`, `L_valor` y `A_valor`, respectivamente. El resultado de esta operación se almacena en la variable `P_valor`.

Es importante notar que la función `subs` no modifica la expresión original, en este caso `P_sol`, sino que devuelve una nueva expresión con las sustituciones realizadas.
